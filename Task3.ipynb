```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("Dataset.csv")
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df,
    x="distance_to_mines_km",
    y="forest_cover_loss_pct",
    hue="high_risk",
    palette={0: "blue", 1: "red"},
    alpha=0.7
)

plt.title("Forest Cover Loss vs Distance to Mines\n(Red = High Risk, Blue = Low Risk)")
plt.xlabel("Distance to Mines (km)")
plt.ylabel("Forest Cover Loss (%)")
plt.legend(title="Risk Level", labels=["High Risk", "Low Risk"])
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df,
    x="reported_illnesses_per1k",
    y="forest_cover_loss_pct",
    hue="high_risk",
    palette={0: "blue", 1: "red"},
    alpha=0.7
)

plt.title("Forest Cover Loss vs Reported Illness Rate\n(Red = High Risk, Blue = Low Risk)")
plt.xlabel("Reported Illnesses (per 1k people)")
plt.ylabel("Forest Cover Loss (%)")
plt.legend(title="Risk Level", labels=["High Risk", "Low Risk"])
plt.grid(True)
plt.tight_layout()
plt.show()
```
